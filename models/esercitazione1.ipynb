```python
#set random seed 
torch.manual_seed(0)

tensorA = torch.rand(7,7)
tensorB = torch.rand(1,7)

Matmul = tensorA @ tensorB.T
print(Matmul)
```

```python
if torch.cuda.is_available():
    torch.cuda.manual_seed(1234) 

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create a random tensor (2, 3) directly on the GPU
tensorA = torch.randn(2, 3, device=device)
tensorB = torch.randn(2, 3, device=device)


matmul = tensorA @ tensorB.T
print(matmul)
```

```python
max = matmul.max()
min = matmul.min()
argmin = matmul.argmin()
argmax = matmul.argmax()

print(max , min, argmin, argmax)
```
